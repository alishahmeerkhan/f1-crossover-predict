```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dotenv import load_dotenv
import os
from pathlib import Path
```

```python
load_dotenv()

file_path = os.getenv('LOAD_PATH')

if not file_path:
    raise ValueError('CRITICAL: File path not found.')
```

```python
df = pd.read_csv(file_path)
```

```python
df.head()
```

## Data Details
To predict -> LapTime_s<br>
How much time will be lost (time delta) give the current situation (features)

```python
fig, axis = plt.subplots(1, 2, figsize=(15, 7))
plt.subplot(121)
sns.barplot(data=df, x='Compound', y='LapTime_s', hue='Rainfall')

plt.subplot(122)
sns.scatterplot(data=df, x='TrackTemp', y='LapTime_s')
plt.show()
```

```python
fig, axis = plt.subplots(1, 3, figsize=(18, 5))

sns.lineplot(data=df, x='TyreLife', y='LapTime_s', ax=axis[0])
axis[0].set_title('Tyre Degradation')


sns.lineplot(data=df, x='AirTemp', y='LapTime_s', ax=axis[1])
axis[1].set_title('Air Temp Vs. Lap Time (s)')


sns.lineplot(data=df, x='TrackTemp', y='LapTime_s', ax=axis[2])
axis[2].set_title('Track Temp Vs. Lap Time (s)')

plt.tight_layout()
plt.show()
```
